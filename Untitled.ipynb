```python
import numpy as np
import itertools
import pandas as pd
import matplotlib.pyplot as plt
import time
```

```python
def calc(array):
    return (sum(array**2))**0.5
```

```python
def dist(array1, array2):
    if len(array1) == len(array2):
        value = calc(array1 - array2)
        return value
    else:
        return "failed"
```

```python
min_distance = "min_distance"
max_distance = "max_distance"
mean_distance = "mean_distance"
median_distance = "median_distance"
distance_std = "distance_std"

array = "array"

stat1 = "stat1"
stat2 = "stat2"
stat3 = "stat3"
```

```python
number = 20
dimensions = 3

iterations = 10**5

D = []
for j in range(iterations):
    arr = np.random.random(size = (number, dimensions))

    combs = itertools.combinations(range(number), 2)
    C = {}
    for i in combs:
        C[i] = dist(arr[i[0]], arr[i[1]])
        
    vals = list(C.values())

    d = {array: arr, "C": C, min_distance: min(vals), 
         max_distance: max(vals), mean_distance: np.mean(vals),
         median_distance: np.median(vals), distance_std: np.std(vals)}

    D.append(d)
    
    if j%(iterations//10) == 0:
        print(j)
    
df = pd.DataFrame(D)
df.loc[:, stat1] = df.loc[:, min_distance] + (df.loc[:, mean_distance])/1.5
df.loc[:, stat2] = df.loc[:, min_distance] - df.loc[:, distance_std]

sort = [stat1, stat2, min_distance, mean_distance, median_distance]
for k in range(len(sort)):
    best = df.sort_values(sort, ascending = False).index

    best_array = df.loc[best[0], array]
    print(best_array)
    for i in sort:
        print(i, df.loc[best[0], i])

    coord = []
    for i in range(dimensions):
        x = np.array([n[i] for n in best_array])
        coord.append(x)

    if dimensions < 3:
        figdim = (3, 3)
        fig, ax = plt.subplots(figsize=figdim, dpi = 150)
        ax.spines["top"].set_visible(False)
        ax.spines["bottom"].set_visible(False)
        ax.spines["right"].set_visible(False)
        ax.spines["left"].set_visible(False)
        x = coord[0]
        if dimensions == 2:
            y = coord[1]
        else:
            y = np.zeros(number)
        plt.scatter(x, y)

        plt.xticks(np.arange(0, 1.2, 0.2))
        plt.yticks(np.arange(0, 1.2, 0.2))
        plt.grid(True)

        plt.show()
    
    elif dimensions in [3, 4]:
        figdim = (3, 3)
        fig, ax = plt.subplots(figsize=figdim, dpi = 150)
        ax.spines["top"].set_visible(False)
        ax.spines["bottom"].set_visible(False)
        ax.spines["right"].set_visible(False)
        ax.spines["left"].set_visible(False)
        
        x_pos = range(number)
        y = [1] * number
        plt.bar(x_pos, y, align = "center", color = best_array)
        plt.title(sort[0])
        
        plt.show()
    
    hum = sort.pop(0)
    sort += [hum]
```

```python
def primes0(size, **kwargs):
    D = {}
    #primes = []
    q = 2
    while q <= size:
        if q not in D:
            yield q
            D[q*q] = [q]
        else:
            for p in D[q]:
                D.setdefault(p + q, []).append(p)
            del D[q]
        q += 1
```

```python
def primes1(size, **kwargs):
    primes = []

    i = 2
    while i <= size:
        if len(primes) == 0:
            primes.append(i)
            i += 1
        else:
            for prime in primes:
                if prime <= i**0.5:
                    if i%prime == 0:
                        i += 2
                        break
                else:
                    primes.append(i)
                    i += 2
                    break
    return primes  
```

```python
s = 10**10**2

start = time.time()
z0 = primes0(s)
print("prime0", time.time() - start)
```

```python
def base(putin, base=2):
    
    i = 0
    while base**(i+1) <= putin:
        i += 1
        
    res = []
    for j in np.arange(i, -1, -1):
        x = 0
        while base**j <= putin:
            putin -= base**j
            x += 1
        res.append(x)
    return res
```

```python
def unbase(putin, base = 2):
    change = [base**n for n in np.arange(len(putin)-1, -1, -1)]
    res = 0
    for i in range(len(putin)):
        res += putin[i] * change[i]
    return res
```

```python
size = 10**7
primes = primes1(size)
```

```python
b = 4
fn = []
for start in primes:
    x1 = base(start, base = b)
    x2 = list(reversed(x1))

    y = unbase(x2, base = b)
    fn.append(start - y)

figdim = (12, 10)
fig, ax = plt.subplots(figsize=figdim)
ax.spines["top"].set_visible(False)
ax.spines["bottom"].set_visible(False)
ax.spines["right"].set_visible(False)
ax.spines["left"].set_visible(False)

plt.scatter(x = primes, y = fn, s = (10**4)/size)

plt.xlabel("primes")
plt.ylabel("f(n)")
plt.grid(True)
plt.show()
```
